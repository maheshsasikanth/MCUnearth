```python
#Get all the stock links from MC & create a DF

import pandas as pd
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://www.moneycontrol.com/india/stockpricequote/"  # Replace with the URL of the website you want to scrape
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find and extract specific data from the website
# Here's an example of extracting all the links on the page

data = []
phrase = "https://www.moneycontrol.com/india/stockpricequote/"
links = soup.find_all('a')
for link in links:
    if link.get('href') is not None:
        if phrase in link.get('href'):
            #print(link.get('href'))
            data.append(link.get('href'))
        else:
            pass
    else:
        pass

df = pd.DataFrame(data, columns=['Stock_Link'])
print(df)

#for index, row in df.iterrows():
 #   print(row['Stock_Link'])
    
```
